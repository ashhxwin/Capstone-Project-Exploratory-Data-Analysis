## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
````

````python
# 1. Remove duplicate rows
initial_shape = df.shape
df = df.drop_duplicates()
print(f"Removed {initial_shape[0] - df.shape[0]} duplicate rows. New shape: {df.shape}")
````

````python
# 2. Handling missing values:

# For 'children' column, fill missing values with 0 (assuming no children)
df['children'].fillna(0, inplace=True)
````

````python
df['country'].fillna('Unknown', inplace=True)
````

````python
df['agent'].fillna(0, inplace=True)
df['company'].fillna(0, inplace=True)
````

````python
# Verify no missing values remain
print("\nMissing values after imputation:")
print(df.isnull().sum())
````

````python
# 3. Convert data types if needed (example: 'agent' and 'company' to int)
df['agent'] = df['agent'].astype(int)
df['company'] = df['company'].astype(int)
````

````python
# 4. Convert 'reservation_status_date' and arrival date components to datetime for easier analysis
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
df['arrival_date'] = pd.to_datetime(df['arrival_date_year'].astype(str) + '-' +
                                    df['arrival_date_month'].astype(str) + '-' +
                                    df['arrival_date_day_of_month'].astype(str))
````

````python
# 5. Check for any anomalous values (example: negative or zero values in lead_time)
print("\nChecking for anomalies:")
print(f"Minimum lead_time: {df['lead_time'].min()}")
print(f"Number of bookings with lead_time <= 0: {(df['lead_time'] <= 0).sum()}")
````

````python
# 6. Final dataset shape after cleaning
print(f"\nFinal dataset shape after cleaning: {df.shape}")
````

### What all manipulations have you done and insights you found?

Removed 31,994 duplicate rows, reducing dataset size to unique bookings.

Imputed missing values:

Filled missing children values with 0.

Filled missing country with 'Unknown' for clarity.

Filled missing agent and company with 0 to indicate absence.

Converted relevant columns to appropriate data types for analysis.

Created a unified arrival_date datetime column for easier time-based analysis.

Dataset is now clean, with no missing values and ready for exploratory analysis.
